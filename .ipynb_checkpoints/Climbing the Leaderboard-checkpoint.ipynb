```python
#循环不是好循环，重新考虑下吧。
def climbingLeaderboard(scores, alice):
    s_rank = [1]*len(scores)
    num = 1
    for i in range(1,len(scores)):
        if scores[i-1] == scores[i]:
            s_rank[i] = num
        elif scores[i-1] > scores[i]:
            num = num +1
            s_rank[i] = num
    print('s_rank\n',s_rank)
    rank = [1]*len(alice)
    rank[0] = s_rank[-1]+1
    start = 1
    record = []
    for i in range(len(alice)):
        for j in range(start,len(scores)):
            if alice[i] == scores[-j]:
                rank[i] = s_rank[-j]
                break
            elif alice[i] < scores[-j]:
                record.append(j)
                rank[i] = s_rank[-j]+1
                start = j
                break
    
    print('record\n',record,'rank\n',rank) 
```

```python
climbingLeaderboard(scores, alice)
```
